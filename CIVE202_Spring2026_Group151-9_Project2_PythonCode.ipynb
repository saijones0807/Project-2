```python
# --- SECTION 1: PRE-VALIDATED NDOT MIX SCENARIOS ---
# These scenarios are based on NDOT Standard Specifications and research data.
# Each dictionary entry contains the target weights (A-D), parameters (E-I), and SGs (J-P).
scenarios = {
    "1": {
        "name": "Class 47B Paving (Standard)",
        "A": 510, "B": 110, "C": 0, "D": 0, "E": 0.44, "F": 6.5, 
        "G": 70, "H": 30, "I": 0, "J": 3.15, "K": 2.30, "L": 2.20, 
        "M": 2.20, "N": 2.62, "O": 2.66, "P": 2.62
    },
    "2": {
        "name": "Class 47BD Bridge Deck (High Performance)",
        "A": 590, "B": 130, "C": 40, "D": 0, "E": 0.40, "F": 7.0, 
        "G": 40, "H": 60, "I": 0, "J": 3.15, "K": 2.30, "L": 2.20, 
        "M": 2.20, "N": 2.62, "O": 2.66, "P": 2.62
    },
    "3": {
        "name": "Class 47BR (Revised/Optimized)",
        "A": 520, "B": 100, "C": 0, "D": 0, "E": 0.40, "F": 6.0, 
        "G": 60, "H": 40, "I": 0, "J": 3.15, "K": 2.30, "L": 2.20, 
        "M": 2.20, "N": 2.62, "O": 2.66, "P": 2.62
    },
    "4": {
        "name": "Class PR3 (High Early Strength)",
        "A": 650, "B": 0, "C": 50, "D": 0, "E": 0.38, "F": 6.5, 
        "G": 45, "H": 55, "I": 0, "J": 3.15, "K": 2.30, "L": 2.20, 
        "M": 2.20, "N": 2.62, "O": 2.66, "P": 2.62
    }
}

# --- SECTION 2: INPUT COLLECTION ---
print("--- NDOT CONCRETE MIX DESIGN AUTOMATOR ---")
print("1. Load Class 47B Paving")
print("2. Load Class 47BD Bridge Deck")
print("3. Load Class 47BR Optimized")
print("4. Load Class PR3 High Early")
print("5. Manual/Custom Input")

choice = input("\nSelect an option (1-5): ")

# Initialize variables for project info
proj_no = "V&V Testing"
control_no = "T-001"

if choice in scenarios:
    # Auto-load the researched data
    s = scenarios[choice]
    print(f"\nSuccessfully loaded: {s['name']}")
    concrete_class = s['name']
    A, B, C, D, E, F, G, H, I = s['A'], s['B'], s['C'], s['D'], s['E'], s['F'], s['G'], s['H'], s['I']
    J, K, L, M, N, O, P = s['J'], s['K'], s['L'], s['M'], s['N'], s['O'], s['P']
else:
    # Sequential User Inputs to mirror Excel Worksheet layout
    print("\n--- Manual Entry Mode ---")
    proj_no = input("Enter Project No: ")
    control_no = input("Enter Control No: ")
    concrete_class = input("Enter Class of Concrete: ")

    print("\n[Step 1: Enter Target Amounts (lb/yd3)]")
    A = float(input("Weight of Cement [A]: "))
    B = float(input("Weight of Fly Ash [B]: "))
    C = float(input("Weight of Silica Fume [C]: "))
    D = float(input("Weight of other SCM [D]: "))

    print("\n[Step 2: Enter Design Parameters]")
    E = float(input("Target Water/Cement Ratio [E]: "))
    F = float(input("Target % Air Content [F]: "))
    G = float(input("Target % Fine Aggregate [G]: "))
    H = float(input("Target % Coarse Aggregate [H]: "))
    I = float(input("Target % Other Aggregate [I]: "))

    print("\n[Step 3: Enter Material Specific Gravities]")
    J = float(input("SG - Cement [J]: "))
    K = float(input("SG - Fly Ash [K]: "))
    L = float(input("SG - Silica Fume [L]: "))
    M = float(input("SG - Other SCM [M]: "))
    N = float(input("SG - Fine Aggregate [N]: "))
    O = float(input("SG - Coarse Aggregate [O]: "))
    P = float(input("SG - Other Aggregate [P]: "))

# --- SECTION 3: CALCULATIONS (Based on NDOT Formulas) ---

# Q: Weight of Water calculation: (A+B+C+D)*E
Q = (A + B + C + D) * E

# Volumetric yields (ft3) based on unit weight of water (62.4 lb/ft3)
# Handling potential divisions by zero with inline checks
R = A / (J * 62.4) if J > 0 else 0
S = B / (K * 62.4) if K > 0 else 0
T = C / (L * 62.4) if L > 0 else 0
U = D / (M * 62.4) if M > 0 else 0
V = (F / 100) * 27 # Volume of Air
W = Q / 62.4      # Volume of Water

# X: Volume of Total Aggregate (Total 27ft3 yard minus other constituents)
X = 27 - R - S - T - U - V - W

# Aggregate Weights (Y-AA)
Y = 62.4 * (G / 100) * N * X # Fine Aggregate
Z = 62.4 * (H / 100) * O * X # Coarse Aggregate
AA = 62.4 * (I / 100) * P * X # Other Aggregate

# --- SECTION 4: OUTPUT GENERATION ---

print("\n" + "="*55)
print("CONTRACTOR TARGET MIX DESIGN WEIGHTS FOR 1 CUBIC YARD")
print(f"PROJECT: {proj_no} | CLASS: {concrete_class}")
print("="*55)
print(f"{'CEMENT:':<25} {A:>15.2f} lb")
print(f"{'FLY ASH:':<25} {B:>15.2f} lb")
print(f"{'SILICA FUME:':<25} {C:>15.2f} lb")
print(f"{'OTHER SCM:':<25} {D:>15.2f} lb")
print(f"{'FINE AGGREGATE:':<25} {Y:>15.2f} lb")
print(f"{'COARSE AGGREGATE:':<25} {Z:>15.2f} lb")
print(f"{'OTHER AGGREGATE:':<25} {AA:>15.2f} lb")
print(f"{'WATER:':<25} {Q:>15.2f} lb")
print("-" * 55)
print(f"{'TOTAL THEORETICAL YIELD:':<25} {R+S+T+U+V+W+X:>15.2f} ft3")
print("="*55)
```
